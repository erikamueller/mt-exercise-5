```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d ={
    "beam_size" : [1,2,3,4,5,6,8,10,12,15],
    "bleu":[19.8,20.6,20.9,21.0,20.9,21.0,21.0,21.0,20.8,20.4]
    }
```

```python
df = pd.DataFrame(data=d)

sns.set_theme(style="darkgrid")
lineplot = sns.lineplot(data=df,x="beam_size",y="bleu")
fig = lineplot.get_figure()
fig.savefig("bleu_beam.png",dpi=300)
```
